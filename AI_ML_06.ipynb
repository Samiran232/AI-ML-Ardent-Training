```python
x=df.iloc[:, :-1]
```

```python
x.shape
```

```python
x
```

```python
y=df.iloc[:, -1]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=41)

#Don't change the values multiple time = random_state=41
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
std=StandardScaler()
mm=MinMaxScaler()
```

```python
x_train_std_sc_data=std.fit_transform(x_train)
x_train_std_sc_data
```

```python
df_std=pd.DataFrame(x_train_std_sc_data,columns=list(c.feature_names))
df_std
```

```python
x_train_std_sc_data
```

```python
x_train_mm_sc_data=mm.fit_transform(x_train)
x_train_mm_sc_data
```

```python
df_mm=pd.DataFrame(x_train_mm_sc_data,columns=list(c.feature_names))
df_mm
```

```python
x_test_std_sc_data=std.transform(x_test)
x_test_std_sc_data
```

```python
x_test_mm_sc_data=mm.transform(x_test)
x_test_mm_sc_data
```

**Reasoning**:
Now that both `x_test_std_sc_data` and `x_test_mm_sc_data` have been created, I will convert `x_test_std_sc_data` into a pandas DataFrame using the original feature names (`c.feature_names`).



```python
df_std_test = pd.DataFrame(x_test_std_sc_data, columns=list(c.feature_names))
df_std_test
```
